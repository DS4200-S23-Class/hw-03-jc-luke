```python
# Plot a pie chart showing proportions of students per programming language

# Get the counts of each value in the Programming_Languages_Most_Familiar_With column
prog_lang_val_counts = df["Programming_Languages_Most_Familiar_With"].value_counts()
#print(prog_lang_val_counts)
#print(" ")

val_proportions = []
for count in prog_lang_val_counts:
    val_proportion = count / sum(prog_lang_val_counts)
    val_proportions.append(val_proportion)
#print(val_proportions)

plt.pie(val_proportions, 
        labels=["Python", "Java", "JavaScript", "R", "Plotly"],
        #pctdistance=0.75,
        labeldistance=1.025,
        shadow=True, 
        startangle=90,
        #autopct="%.1f%%",
        wedgeprops={"edgecolor": "black"},
        textprops={"color": "black", "fontsize": 7},
        rotatelabels=True)

plt.title("Proportion of Students in DS4200 Based on Programming Language Most Familiar With")
plt.tight_layout()

plt.show()
```
